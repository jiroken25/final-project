```python
url = "https://covid-19-data.p.rapidapi.com/country/all"

querystring = {"format":"json"}

headers = {
    'x-rapidapi-key': key,
    'x-rapidapi-host': "covid-19-data.p.rapidapi.com"
    }

response_covid = requests.request("GET", url, headers=headers, params=querystring)

covid_dict = json.loads(response_covid.text)
```

```python
covid_dict
```

```python
country = []
death = []
lon = []
lat = []

for each in covid_dict:
    country.append(each['country'])
    death.append(each["deaths"])
    lat.append(each["latitude"])
    lon.append(each["longitude"])
    
```

```python
covid_df = pd.DataFrame({"Country":country,"Number of Death":death,"lat":lat,"lon":lon})
```

```python
covid_df_bycode = pd.merge(covid_df,country_searchtable,how="left",left_on="Country",right_on="Country_y")
```

```python
population_src = pd.read_csv("resources/WPP2019_TotalPopulationBySex.csv")
```

```python
population = population_src[population_src.Variant == "Medium"][["LocID", "Location", "Time", "PopTotal"]]
population["PopTotal"] = (population["PopTotal"] * 1000).astype(int)
population = population[population["Time"]==2019]
```

```python
locations_src = pd.read_excel('resources/WPP2019_F01_LOCATIONS.XLSX', sheet_name="DB")
```

```python
location = locations_src[locations_src.LocType == 4][["LocID","ISO3_Code"]]
```

```python
population_bycode = pd.merge(population,location,how="left",left_on=["LocID"], right_on=["LocID"])[["ISO3_Code","PopTotal"]].dropna()
```

```python
population_bycode
```

```python
covid_pop_df = pd.merge(covid_df_bycode,population_bycode,how="left", left_on=["CODE"], right_on=["ISO3_Code"])
```

```python
covid_pop_df = covid_pop_df.dropna()
```

```python
covid_pop_df.to_csv("covid_pop.csv")
```

```python
covid_pop_df[["CODE","Number of Death","PopTotal","lat","lon"]]
```

```python
covid_db = pd.merge(covid_pop_df,country_code, how="inner", left_on=["CODE"], right_on=["CODE"])[["index_y","Number of Death","PopTotal","lat","lon"]]
covid_db.columns = ["country_id","Death","Population","lat","lon"]
```

```python
covid_db["death/population"] = covid_db["Death"]/ covid_db["Population"]
```

```python
covid_db
```

```python
covid_db["death/population"].describe()
```

```python
# Definition : IF death/population < 0.0001, it's successful

def successful(x):
    if x < 0.0001:
        return "successful"
    else:
        return "not successful"

```

```python
covid_db["success"] = covid_db["death/population"].apply(successful)
```

```python
covid_db
```

```python
covid_db.to_sql("covid_data",con=engine,if_exists='replace',index=False)
```
