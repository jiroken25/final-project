```python
import pandas as pd
```

```python
health_spending = pd.read_csv("resources\health_spending.csv")
```

```python
health_table = pd.merge(health_spending[["LOCATION","TIME"]].groupby("LOCATION").max().reset_index(),health_spending, how="left", left_on=["LOCATION","TIME"], right_on=["LOCATION","TIME"])
```

```python
health_spending_data = health_table[(health_table["SUBJECT"]=="TOT") & (health_table["MEASURE"]=="USD_CAP")].reset_index().drop(["index","Flag Codes"], axis=1)
```

```python
health_spending_data
```
