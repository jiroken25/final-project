```python
import pandas as pd
```

```python
health_spending = pd.read_csv("resources\health_spending.csv")
```

```python
health_table = pd.merge(health_spending[["LOCATION","TIME"]].groupby("LOCATION").max().reset_index(),health_spending, how="left", left_on=["LOCATION","TIME"], right_on=["LOCATION","TIME"])
```

```python
health_spending_data = health_table[(health_table["SUBJECT"]=="TOT") & (health_table["MEASURE"]=="USD_CAP")].reset_index().drop(["index","Flag Codes"], axis=1)
```

```python
health_spending_data
```

```python
hospital_bed = pd.read_csv("resources\hospital_bed.csv")
```

```python
hos_bed_table = pd.merge(hospital_bed[["LOCATION","TIME"]].groupby("LOCATION").max().reset_index(),hospital_bed, how="left", left_on=["LOCATION","TIME"], right_on=["LOCATION","TIME"])
```

```python
unemployment =  pd.read_csv("resources\longterm_unemployment.csv")
unemployment_table = pd.merge(unemployment[["LOCATION","TIME"]].groupby("LOCATION").max().reset_index(),unemployment, how="left", left_on=["LOCATION","TIME"], right_on=["LOCATION","TIME"])
unemployment_data = unemployment_table[unemployment_table["SUBJECT"]=="TOT"].reset_index().drop(["index","Flag Codes"], axis=1)
unemployment_data
```

```python
obesity_table = pd.merge(obesity[["LOCATION","TIME"]].groupby("LOCATION").max().reset_index(),obesity, how="left", left_on=["LOCATION","TIME"], right_on=["LOCATION","TIME"])
obesity_data = obesity_table.reset_index().drop(["index","Flag Codes"], axis=1)
# average of measured data and selt reported data
obesity_data = obesity_data.groupby("LOCATION").mean().reset_index()
obesity_data
```
